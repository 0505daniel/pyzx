```python
import sys
import random
sys.path.append('..')
import pyzx as zx
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
g = zx.examples.cnots(5, 10**5)
```

```python
zx.simplify.phase_free_simp(g)
```

```python
zx.drawing.pack_circuit_nf(g,'grg')
zx.draw(g)
```

```python
random.seed(99)
g = zx.cnots(5, 50, backend='igraph')
zx.draw(g)
```

```python
zx.phase_free_simp(g)
zx.drawing.pack_circuit_nf(g, 'grg')
zx.draw(g)
```

```python
%time g = zx.cnots(5, 10**6, backend='igraph')
print('built CNOT graph with ' + str(g.num_vertices()) + ' vertices')
%time zx.phase_free_simp(g)
print(g.num_vertices())
print('final graph has ' + str(g.num_vertices()) + ' vertices')
```

```python
zx.drawing.pack_circuit_nf(g, 'grg')
zx.draw(g)
```
